```python
import numpy as np
import pandas as pd
```

## Zona Federal

```python
def get_lon(anexos):
    #-d - m / 60 - s / 3600 
    return -anexos['GRADOS LONGITUD'] - anexos['MINUTOS LONGITUD']/60 - anexos['SEGUNDOS LONGITUD']/3600
```

```python
def get_lat(anexos):
    #d + m / 60 + s / 3600 
    return anexos['GRADOS LONGITUD'] + anexos['MINUTOS LONGITUD']/60 + anexos['SEGUNDOS LONGITUD']/3600
```

```python
anexosZon.columns
```

```python
zonaFederal = anexosZon[['TÍTULO','NÚMERO DE ANEXO', 'SUPERFICIE EN m2', 'CORRIENTE O VASO','CLAVE DE MUNICIPIO','ESTADO','CLAVE DE CUENCA HIDROLÓGICA']]
```

```python
zonaFederal['NÚMERO DE ANEXO'] = np.arange(zonaFederal.shape[0])
zonaFederal['LONIGTUD'] = get_lon(anexosZon).copy()
zonaFederal['LATITUD'] = get_lat(anexosZon).copy()
```

```python
zonaFederal
```

```python
zonaFederal.to_csv(dfolder+'zonaFed.csv', index=False)
```

## Anexo de extracción

```python
supers = anexosSup[['TÍTULO', 'NÚMERO DE ANEXO']]
supers['LONGITUD'] =  get_lon(anexosSup)
supers['LATITUD'] = get_lon(anexosSup)
supers['TIPO'] = ['SUPERFICIAL']*supers.shape[0]
```

```python
subtes = anexosSub[['TÍTULO', 'NÚMERO DE ANEXO']]
subtes['LONGITUD'] =  get_lon(anexosSub)
subtes['LATITUD'] = get_lon(anexosSub)
subtes['TIPO'] = ['SUBTERRÁNEO']*subtes.shape[0]
```

```python
anexosExt = pd.concat([supers,subtes])
anexosExt['NÚMERO DE ANEXO'] = np.arange(anexosExt.shape[0])
anexosExt
```

```python
anexosExt.to_csv(dfolder+'anexoExt.csv')
```

## Anexo-acuifero

```python
acs_homo = anexosExt[anexosExt.TIPO=='SUBTERRÁNEO'].join(anexosSub[['CLAVE DE ACUÍFERO HOMOLOGADO','VOLUMEN ANUAL EN m3']])[['NÚMERO DE ANEXO','CLAVE DE ACUÍFERO HOMOLOGADO','VOLUMEN ANUAL EN m3']]
acs_homo = acs_homo.rename(columns={'CLAVE DE ACUÍFERO HOMOLOGADO':'CLAVE DE ACUÍFERO'})

acs_men = anexosExt[anexosExt.TIPO=='SUBTERRÁNEO'].join(anexosSub[['CLAVE DE ACUÍFERO QUE MENCIONA EL TÍTULO','VOLUMEN ANUAL EN m3']])[['NÚMERO DE ANEXO','CLAVE DE ACUÍFERO QUE MENCIONA EL TÍTULO','VOLUMEN ANUAL EN m3']]
acs_men = acs_men.rename(columns={'CLAVE DE ACUÍFERO QUE MENCIONA EL TÍTULO':'CLAVE DE ACUÍFERO'})
```

```python
anexoExt_acs = pd.concat([acs_homo,acs_men])
anexoExt_acs = anexoExt_acs[anexoExt_acs['CLAVE DE ACUÍFERO']!=0]
anexoExt_acs.drop_duplicates(inplace=True)
anexoExt_acs['FECHA']=['2022-05-22']*anexoExt_acs.shape[0]
anexoExt_acs
```

```python
anexoExt_acs.to_csv(dfolder+'anexoExt-ac.csv',index=False)
```
